```python
from ttsmms import TTS
```

```python
tts=TTS("data/mya") #update lang code
```

Short-Text

```python
wav=tts.synthesis("မြန်မာလို စမ်းပြောကြည့်မယ်")
```

```python
from IPython.display import Audio
Audio(wav["x"], rate=wav["sampling_rate"])
```

Slightly Long Text

```python
text = """မြန်မာလို စမ်းပြောကြည့်တာ အသံက သိပ်အဆင်မပြေသေးဘူး။ ဒီတိုင်းဆို သုံးလို့ အဆင်မပြေသေးပါဘူး"""
```

```python
wav_text =tts.synthesis(text)
```

```python
from IPython.display import Audio
Audio(wav_text["x"], rate=wav["sampling_rate"])
```
